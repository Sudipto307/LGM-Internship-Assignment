```python
import cv2
import matplotlib.pyplot as plt
```

### Loading Original Image

```python
image = cv2.imread("bird.jpg")
image_1 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_1)
plt.axis('off')
plt.title('Original_image')
plt.show()
```

### Converting Image to Gray Scale

```python
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(image_gray,cmap='gray')
plt.axis('off')
plt.title('Gray Scale Image')
plt.show()
```

### Inverting The Gray Scaled Image

```python
image_invert = cv2.bitwise_not(image_gray)
plt.figure(figsize=(8,8))
plt.imshow(image_invert,cmap='gray')
plt.axis('off')
plt.title('Inverted Image')
plt.show()
```

### Bluring the Image using Gaussian Blur

```python
image_blur = cv2.GaussianBlur(image_invert,(21,21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(image_blur ,cmap='gray')
plt.axis('off')
plt.title('Blurred image')
plt.show()
```

### Getting The Pencil Sketch of the Image

```python
image_pencil = cv2.divide(image_gray,255-image_blur, scale = 255)
plt.figure(figsize=(8,8))
plt.imshow(image_pencil,cmap='gray')
plt.axis('off')
plt.title('Fianl Pencile Sketch')
plt.show()
```
