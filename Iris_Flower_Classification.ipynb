```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('iris.data', header = None)
df.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm', 'PetalWidthCm', 'Class']
df
```

```python
df['Class'].unique()
```

## Visualizations

```python
plt.figure(figsize = (10,6))
ax1 = sns.scatterplot(x = df['SepalLengthCm'], y = df['SepalWidthCm'], hue = df['Class'])
plt.legend(loc = 'upper right')
ax1.set_xlabel('Sepal Length in cm', fontsize = 18)
ax1.set_ylabel('Sepal Width in cm', fontsize = 18)
ax1.set_title('Sepal Length Vs Sepal Width for Different Classes', fontsize = 22)
```

```python
plt.figure(figsize = (10,6))
ax2 = sns.scatterplot(x = df['PetalLengthCm'], y = df['PetalWidthCm'], hue = df['Class'])
plt.legend(loc = 'lower right')
ax2.set_xlabel('Petal Length in cm', fontsize = 18)
ax2.set_ylabel('Petal Width in cm', fontsize = 18)
ax2.set_title('Petal Length Vs Petal Width for Different Classes', fontsize = 22)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state = 0)
```

## Modelling

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression().fit(X_train, y_train)
yhat1 = model1.predict(X_test)
score1 = accuracy_score(yhat1, y_test)
score1
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
k = []
score = []
for i in range(1,10):
    model2 = KNeighborsClassifier(n_neighbors = i).fit(X_train, y_train)
    yhat2 = model2.predict(X_test)
    score2 = accuracy_score(yhat2, y_test)
    k.append(i)
    score.append( score2 )
plt.plot(k,score)
score
```

### SVM

```python
from sklearn.svm import SVC
model3 = SVC().fit(X_train, y_train)
yhat3 = model3.predict(X_test)
score3 = accuracy_score(yhat3, y_test)
score3
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model4 = DecisionTreeClassifier().fit(X_train, y_train)
yhat4 = model4.predict(X_test)
score4 = accuracy_score(yhat4, y_test)
score4
```

## Using different classifiers like logistic regression, K Nearest  Neighbours, Support Vector Classifier and Decission Tree , the accuracy score is  remaining same

## So we can use any classifier to make the classification for this dataset
